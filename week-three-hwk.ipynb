Suppose you have ridden a bicycle from New York City to Key West, Florida.  Your bicycle odometer shows the total miles, which you have dutifully logged at the end of each day with paper and pencil.  Your first two entries might be 55 and 120, indicating that you rode your bike 55 miles on day 1 and 65 miles on day 2.  Your task is to create a pandas Series object that holds the cumulative miles at the end of each day, then use your Python skills to show the total miles that you rode each day.  Consider how you should best present this information in a Jupyter notebook.  You should save your Jupyter Notebook to your GitHub repository, and provide a link in your assignment submission. 

Assumptions: According to a quick google search; traveling between NYC and Key West is approximately 1,440 road miles. We'll use a the example of 55 miles on day 1 and 65 miles on day 2 as the minimum and maximum boundaries for the number of miles traveled per day. 

```python
import random
import pandas as pd

def generate_bike_odometer_series(max_value=1440):
    total = 0
    values = []
    
    while total < max_value:
        remaining = max_value - total
        
        # If remaining is less than 55, use the remaining value to reach the max value
        if remaining < 55:
            values.append(max_value)
            total = max_value
        else:
            # Generate a random number between 55 and 65 representing the distance traveled in a day
            num = random.randint(55, 65)
            
            # If adding the random number would exceed max_value,
            # just add the remaining amount instead
            if total + num > max_value:
                values.append(max_value)
                total = max_value
            else:
                total += num
                values.append(total)
    
    return pd.Series(values)

trip_log_series = generate_bike_odometer_series(300)
print("The generated series is: {}".format(trip_log_series))
```

Suppose you have ridden a bicycle from New York City to Key West, Florida.  Your bicycle odometer shows the total miles, which you have dutifully logged at the end of each day with paper and pencil.  Your first two entries might be 55 and 120, indicating that you rode your bike 55 miles on day 1 and 65 miles on day 2.  Your task is to create a pandas Series object that holds the cumulative miles at the end of each day, then use your Python skills to show the total miles that you rode each day. Assumptions: According to a quick google search; traveling between NYC and Key West is approximately 1,440 road miles. We'll use a the example of 55 miles on day 1 and 65 miles on day 2 as the minimum and maximum boundaries for the number of miles traveled per day. 

Generate a pandas Series object that holds the cumulative miles at the end of each days. The trip from NYC to Key West, FL is approximately 1440 road miles, and the number of miles per day will range between 55 and 65 miles. The total miles traveled will not exceed 1440 miles. The resulted series is printed at the end. 

```python
import random
import pandas as pd

def generate_bike_odometer_series(max_value=1440):
    total = 0
    values = []
    
    while total < max_value:
        remaining = max_value - total
        
        # If remaining is less than 55, use the remaining value to reach the max value
        if remaining < 55:
            values.append(max_value)
            total = max_value
        else:
            # Generate a random number between 55 and 65 representing the distance traveled in a day
            num = random.randint(55, 65)
            
            # If adding the random number would exceed max_value,
            # just add the remaining amount instead
            if total + num > max_value:
                values.append(max_value)
                total = max_value
            else:
                total += num
                values.append(total)
    
    return pd.Series(values)


trip_log_series = generate_bike_odometer_series(1440)
print("The generated series is: \n{}".format(trip_log_series))
```

The function calculates the number of miles traveled per day and prints it out as it iterates through the odometer series. 

```python
import random
import pandas as pd

# using a recursive function to print the daily travel for fun/practice
def print_dialy_travel(odo_series, index=None):
    # Initialize index on first call
    if index is None:
        index = len(odo_series) - 1

    # Base case exits the function after iterating through the entire series
    if index <= 0:
        return "finished printing data"

    # Calculate then print the difference between current and previous day in the log
    difference = odo_series.iloc[index] - odo_series.iloc[index - 1]
    print("Daily travel on day {}: {} miles".format(index, difference))

    # Recursively call the function using the next index value until base case is reached
    return print_dialy_travel(odo_series, index - 1)


print_dialy_travel(trip_log_series)
```
